```python
import requests
import pandas as pd
import time

session = requests.Session()

df_list = []

for i in range(1, 531):
    try:
        url = f"https://api.themoviedb.org/3/movie/top_rated?api_key=77b5f7b284f69b451939b65ceadc38fc&language=en-US&page={i}"
        
        response = session.get(url, timeout=10)
        response.raise_for_status()

        temp_df = pd.DataFrame(response.json()['results'])[
            ['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']
        ]

        df_list.append(temp_df)
        print(f"Page {i} fetched ✅")

        time.sleep(0.5)  # ⛔ VERY IMPORTANT (rate limit)

    except Exception as e:
        print(f"Page {i} skipped ❌ | Reason: {e}")
        time.sleep(1)
        continue

df = pd.concat(df_list, ignore_index=True)
```
